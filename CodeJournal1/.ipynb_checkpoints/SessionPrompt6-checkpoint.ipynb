```python
import numpy as np
```

Sinfunc is a function that returns sin(x).

```python
def sinfunc(x): #defines the main function
	return np.sin(x) 
```

Cosfunc is a function that returns cos(x).

```python
def cosfunc(x): #defines the main function
	return np.cos(x) 
```

```python
from tabulate import tabulate #python built-in package to make tables

twopi = 2 * np.pi
x = np.linspace(0, twopi, 1000)
y = sinfunc(x)

table_data = [(a,b) for a,b in zip(x,y)]

table_headers = ["x","sin(x)"]
python_table = tabulate(table_data, tablefmt="grid", headers=table_headers, floatfmt=".3f")

y2 = cosfunc(x)

table_data2 = [(a,b) for a,b in zip(x,y2)]

table_headers2 = ["x","cos(x)"]
python_table2 = tabulate(table_data2, tablefmt="grid", headers=table_headers2, floatfmt=".3f")

print(python_table)
print(python_table2)

for i in range(10):
    print(table_data[i][0], table_data[i][1], table_data2[i][1])


```

```python
twopi = 2 * np.pi
x = np.linspace(0, twopi, 1000)
y = sinfunc(x)
z = cosfunc(x)

table_data = [(a,b,c) for a,b,c in zip(x,y,z)]

table_headers = ["x","sin(x)","cos(x)"]
python_table = tabulate(table_data, tablefmt="grid", headers=table_headers, floatfmt=".3f")

print("    x   cosx   sinx")
print("-------------------")
for i in range(10):
    a, b, c = table_data[i]
    print(f"{a:.3f}  {b:.3f}  {c:.3f}")
```
