```python
import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline
```

```python
x = np.linspace(0,np.pi,100)
```

```python
def sin1(x):
    return np.sin(x)
def cos1(x):
    return np.cos(x)
```

```python
sinx = sin1(x)
cosx = cos1(x)
```

```python
figure, axes = plt.subplots(1,2)
axes[0].plot(x, sinx)
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")
axes[0].set_title(r'$\sin(x)$')

axes[1].plot(x, cosx)
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")
axes[1].set_title(r'$\cos(x)$')

figure.subplots_adjust(wspace = 0.5)

plt.savefig('Session8.pdf',bbox_inches='tight',dpi=600)
plt.show()
```
