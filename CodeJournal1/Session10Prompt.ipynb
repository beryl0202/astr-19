```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.power(5, size=1000)

# create the histrogram bins
width = 0.01

bins = np.arange(0,1+width,width) #this array contains the borders of the bins

# create the histogram
plt.hist(x,bins=bins,alpha=0.5,edgecolor='black')
plt.xlabel('x')
plt.ylabel('Counts')

plt.savefig('Session10.pdf',bbox_inches='tight',dpi=600)
plt.show()
```
